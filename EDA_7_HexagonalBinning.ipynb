```python
import pandas as pd
kcData = pd.read_csv('./data/kc_house_data.csv',delimiter=',',encoding='utf-8')
kcData
print(kcData.columns)
```

```python
## Create Scatterplott of sqft & price
data = kcData[['sqft_living', 'price']].values.reshape(-1,2)
data = pd.DataFrame(data, columns = ['sqft_living', 'price']) 
print(data)
```

```python
xdata=data['sqft_living']
ydata=data['price']
```

## Scatterplot f√ºr KingsCounty

```python
ax1 = data.plot.scatter(x='sqft_living',
                      y='price',
                      c='DarkBlue')
```

Das Scatterplott erm√∂glicht keinen konkreten Einblick in einen m√∂glichen Trend

## HexagonnalBinning f√ºr KingsCounty

```python
ax1 = data.plot.hexbin(x='sqft_living',
                      y='price',
                       gridsize=(15,15))

```
