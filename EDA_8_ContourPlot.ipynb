# ContourPlot

```python
import pandas as pd
kcData = pd.read_csv('./data/kc_house_data.csv',delimiter=',',encoding='utf-8')
kcData
print(kcData.columns)
```

```python
## Create Scatterplott of sqft & price
data = kcData[['sqft_living', 'price']].values.reshape(-1,2)
data = pd.DataFrame(data, columns = ['sqft_living', 'price']) 
print(data)
```

```python
xlist=data['sqft_living']
ylist=data['price']
xlist = xlist.loc[:999]
ylist = ylist.loc[:999]
xlist = pd.DataFrame(xlist)
ylist = pd.DataFrame(ylist)
print(xlist.shape)
print(ylist.shape)
```

## Contour Plots
Brauch ich noch nen sinnvollen Anwendungsfall f√ºr

```python
import numpy as np
import matplotlib.pyplot as plt
#xlist = np.linspace(-3.0, 3.0, 100)
#ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
#ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
```
