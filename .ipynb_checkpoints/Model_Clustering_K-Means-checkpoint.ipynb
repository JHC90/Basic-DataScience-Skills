```python
## Dependencies & Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

## Create synthetischen Datensatz
hierdurch weis ich dass es 4 Center geben wird, in normalen Clusterfragestellungen weis ich sowas nicht. 

```python
X, y = make_blobs(n_samples = 150, centers=4, cluster_std = 0.5, random_state = 0)
plt.scatter(X[:, 0], X[:, 1], s=50)
```

## Modelling

```python
# from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 4, random_state=0)
kmeans.fit(X)
```

## Prediciton

### Prediction Training-Dataset

```python
y_pred = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c = y_pred, s=50)
```

### Prediction neue Werte
wenn ich nun von einem neuen Datenpunkt die Gruppe wissen will, so kann ich diese mir detektieren lassen. Hier detektiere ich die Punkte zu der Liste. hier habe ich versucht über meine manuelle auswahl der zu preditetenden Punkte jeweils jedes Cluster einmal zu treffen.

```python
kmeans.predict([[-2, 8], [-2, -1], [1, 4], [2, 0]])
```

## Paramter From Model

### Position Zentroide vom Modell

```python
kmeans.cluster_centers_
```

### Labels
welcher Punkt wird welcher Gruppe zugeordnet

```python
kmeans.labels_
```

### Iterations
wieviele Iterationen hat das Modell gebraucht bis zum stop

```python
kmeans.n_iter_
```
