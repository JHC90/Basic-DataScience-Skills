```python
import pandas as pd
iris = pd.read_csv('./Sample-Projects/Iris/data/iris.data',delimiter=',',encoding='utf-8')
iris.columns = ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']
```

## Korrelations-Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 8))
df_corr = iris.corr()
ax = sns.heatmap(df_corr, annot=True, cmap="RdYlGn") 
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
```

#### Scatter Matrix
In der Scatter matrix werden nur die numerischen Werte eingebaut. An der diagonalen laufen dann die Histogramme der jeweiligen Variablen. 

```python
#!! ! achtung hier werden alle arten gemeinsam dargestellt => frage der Sinnhaftigkeit...check SNS
from pandas.plotting import scatter_matrix
scatter_matrix(iris)
plt.show()
```
